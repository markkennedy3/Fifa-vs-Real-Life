# Accuracy of Attributes

```python
import os.path
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pylab as plt
from matplotlib.gridspec import GridSpec
import matplotlib.gridspec as gridspec
%matplotlib inline 
```

## Attacking Attributes vs Goalscoring Abilities

```python
RQ3_1 = pd.read_csv(('../../data/analysis/RQ3_1.csv'))
RQ3_1 = RQ3_1.drop('Unnamed: 0', 1)
RQ3_1 = RQ3_1.iloc[1:]
```

```python
RQ3_1.head(5)
```

```python
sho = RQ3_1.groupby(['sho']).mean()['Average Goals/Game']
sho.plot(style='.-')
```

```python
sho.corr(RQ3_1['Average Goals/Game'])
```

```python
composure = RQ3_1.groupby(['composure']).mean()['Average Goals/Game']
finishing = RQ3_1.groupby(['finishing']).mean()['Average Goals/Game']
volleys = RQ3_1.groupby(['volleys']).mean()['Average Goals/Game']
shot_power = RQ3_1.groupby(['shot_power']).mean()['Average Goals/Game']
long_shots = RQ3_1.groupby(['long_shots']).mean()['Average Goals/Game']
penalties = RQ3_1.groupby(['penalties']).mean()['Average Goals/Game']
positioning = RQ3_1.groupby(['positioning']).mean()['Average Goals/Game']
heading_accuracy = RQ3_1.groupby(['positioning']).mean()['Average Goals/Game']
```

```python
composure_corr = composure.corr(RQ3_1['Average Goals/Game']).round(5).astype(str) 
finishing_corr = finishing.corr(RQ3_1['Average Goals/Game']).round(5).astype(str) 
volleys_corr = volleys.corr(RQ3_1['Average Goals/Game']).round(5).astype(str) 
shot_power_corr = shot_power.corr(RQ3_1['Average Goals/Game']).round(5).astype(str) 
long_shots_corr = long_shots.corr(RQ3_1['Average Goals/Game']).round(5).astype(str) 
penalties_corr = penalties.corr(RQ3_1['Average Goals/Game']).round(5).astype(str) 
positioning_corr = positioning.corr(RQ3_1['Average Goals/Game']).round(5).astype(str) 
heading_accuracy_corr = heading_accuracy.corr(RQ3_1['Average Goals/Game']).round(5).astype(str) 
```

```python
axis_size=14
color = ['r']

fig = plt.figure(figsize=(30,15))

gs = gridspec.GridSpec(2,6)
ax0 = plt.subplot(gs[0,0:2],title='\n \n'+ 'Correlation: ' + finishing_corr)
ax0.title.set_fontsize(20)
ax0.set_ylabel('Average Goals/Game',size=axis_size)
ax0.set_xlabel('Average Goals/Game',size=axis_size)
finishing.plot(ax=ax0,kind='line',color=color)
ax1 = plt.subplot(gs[0,2:4],title='\n \n'+ 'Correlation: ' + heading_accuracy_corr)
ax1.title.set_fontsize(20)
ax1.set_ylabel('Average Goals/Game',size=axis_size)
ax1.set_xlabel('Average Goals/Game',size=axis_size)
heading_accuracy.plot(ax=ax1,kind='line',color=color)

fig = plt.figure(figsize=(30,15))
gs = gridspec.GridSpec(2,6)
ax0 = plt.subplot(gs[0,0:2],title='\n \n'+ 'Correlation: ' + composure_corr)
ax0.title.set_fontsize(20)
ax0.set_ylabel('Average Goals/Game',size=axis_size)
ax0.set_xlabel('Average Goals/Game',size=axis_size)

composure.plot(ax=ax0,kind='line',color=color)
ax1 = plt.subplot(gs[0,2:4],title='\n \n'+ 'Correlation: ' + volleys_corr)
ax1.title.set_fontsize(20)
ax1.set_ylabel('Average Goals/Game',size=axis_size)
ax1.set_xlabel('Average Goals/Game',size=axis_size)
volleys.plot(ax=ax1,kind='line',color=color)

fig = plt.figure(figsize=(30,15))
gs = gridspec.GridSpec(2,6)
ax0 = plt.subplot(gs[0,0:2],title='\n \n'+ 'Correlation: ' + shot_power_corr)
ax0.title.set_fontsize(20)
ax0.set_ylabel('Average Goals/Game',size=axis_size)
ax0.set_xlabel('Average Goals/Game',size=axis_size)

shot_power.plot(ax=ax0,kind='line',color=color)
ax1 = plt.subplot(gs[0,2:4],title='\n \n'+ 'Correlation: ' + long_shots_corr)
ax1.title.set_fontsize(20)
ax1.set_ylabel('Average Goals/Game',size=axis_size)
ax1.set_xlabel('Average Goals/Game',size=axis_size)
long_shots.plot(ax=ax1,kind='line',color=color)

fig = plt.figure(figsize=(30,15))
gs = gridspec.GridSpec(2,6)
ax0 = plt.subplot(gs[0,0:2],title='\n \n'+ 'Correlation: ' + penalties_corr)
ax0.title.set_fontsize(20)
ax0.set_ylabel('Average Goals/Game',size=axis_size)
ax0.set_xlabel('Average Goals/Game',size=axis_size)

penalties.plot(ax=ax0,kind='line',color=color)
ax1 = plt.subplot(gs[0,2:4],title='\n \n'+ 'Correlation: ' + positioning_corr)
ax1.title.set_fontsize(20)
ax1.set_ylabel('Average Goals/Game',size=axis_size)
ax1.set_xlabel('Average Goals/Game',size=axis_size)
positioning.plot(ax=ax1,kind='line',color=color)
```

## Ability to Assist

```python
RQ3_2 = pd.read_csv(('../../data/analysis/RQ3_2.csv'))
RQ3_2 = RQ3_2.drop('Unnamed: 0', 1)
RQ3_2 = RQ3_2.iloc[1:]
```

```python
pas = RQ3_2.groupby(['pas']).mean()['Average Assists/Game']
pas.plot(style='.-')
```

```python
pas.corr(RQ3_2['Average Assists/Game'])
```

```python
crossing = RQ3_2.groupby(['crossing']).mean()['Average Assists/Game']
vision = RQ3_2.groupby(['vision']).mean()['Average Assists/Game']
short_passing = RQ3_2.groupby(['short_passing']).mean()['Average Assists/Game']
long_passing = RQ3_2.groupby(['long_passing']).mean()['Average Assists/Game']
dribbling = RQ3_2.groupby(['dribbling']).mean()['Average Assists/Game']
```

```python
crossing.head(5)
```

```python
crossing_corr = crossing.corr(RQ3_2['Average Assists/Game']).round(5).astype(str) 
vision_corr = vision.corr(RQ3_2['Average Assists/Game']).round(5).astype(str) 
short_passing_corr = short_passing.corr(RQ3_2['Average Assists/Game']).round(5).astype(str) 
long_passing_corr = long_passing.corr(RQ3_2['Average Assists/Game']).round(5).astype(str) 
dribbling_corr = dribbling.corr(RQ3_2['Average Assists/Game']).round(5).astype(str) 
```

```python
axis_size=14
color = ['r']

fig = plt.figure(figsize=(30,15))

gs = gridspec.GridSpec(2,6)
ax0 = plt.subplot(gs[0,0:2],title='\n \n'+ 'Correlation: ' + crossing_corr)
ax0.title.set_fontsize(20)
ax0.set_ylabel('Average Assists/Game',size=axis_size)
ax0.set_xlabel('Average Assists/Game',size=axis_size)
crossing.plot(ax=ax0,kind='line',color=color)
ax1 = plt.subplot(gs[0,2:4],title='\n \n'+ 'Correlation: ' + vision_corr)
ax1.title.set_fontsize(20)
ax1.set_ylabel('Average Assists/Game',size=axis_size)
ax1.set_xlabel('Average Assists/Game',size=axis_size)
vision.plot(ax=ax1,kind='line',color=color)

fig = plt.figure(figsize=(30,15))
gs = gridspec.GridSpec(2,6)
ax0 = plt.subplot(gs[0,0:2],title='\n \n'+ 'Correlation: ' + short_passing_corr)
ax0.title.set_fontsize(20)
ax0.set_ylabel('Average Assists/Game',size=axis_size)
ax0.set_xlabel('Average Assists/Game',size=axis_size)

short_passing.plot(ax=ax0,kind='line',color=color)
ax1 = plt.subplot(gs[0,2:4],title='\n \n'+ 'Correlation: ' + long_passing_corr)
ax1.title.set_fontsize(20)
ax1.set_ylabel('Average Assists/Game',size=axis_size)
ax1.set_xlabel('Average Assists/Game',size=axis_size)
long_passing.plot(ax=ax1,kind='line',color=color)

fig = plt.figure(figsize=(54,10))
gs = gridspec.GridSpec(2,6)
ax0 = plt.subplot(gs[0,0:2],title='\n \n'+ 'Correlation: ' + dribbling_corr)
ax0.title.set_fontsize(20)
ax0.set_ylabel('Average Assists/Game',size=axis_size)
ax0.set_xlabel('Average Assists/Game',size=axis_size)
dribbling.plot(ax=ax0,kind='line',color=color)
```

## Goalkeeping Attributes vs Ability To Keep Clean Sheets

```python
RQ3_3 = pd.read_csv(('../../data/analysis/RQ3_3.csv'))
RQ3_3 = RQ3_3.drop('Unnamed: 0', 1)
RQ3_3 = RQ3_3.iloc[1:]
```

```python
RQ3_3
```

```python
gk_diving = RQ3_3.groupby(['gk_diving']).mean()['Average Clean Sheets/Game']
gk_handling = RQ3_3.groupby(['gk_handling']).mean()['Average Clean Sheets/Game']
gk_kicking = RQ3_3.groupby(['gk_kicking']).mean()['Average Clean Sheets/Game']
gk_positioning = RQ3_3.groupby(['gk_positioning']).mean()['Average Clean Sheets/Game']
gk_reflexes = RQ3_3.groupby(['gk_reflexes']).mean()['Average Clean Sheets/Game']
```

```python
gk_handling_corr = gk_handling.corr(RQ3_3['Average Clean Sheets/Game']).round(5).astype(str)
gk_kicking_corr = gk_kicking.corr(RQ3_3['Average Clean Sheets/Game']).round(5).astype(str)
gk_positioning_corr = gk_positioning.corr(RQ3_3['Average Clean Sheets/Game']).round(5).astype(str)
gk_reflexes_corr = gk_reflexes.corr(RQ3_3['Average Clean Sheets/Game']).round(5).astype(str)
gk_diving_corr = gk_diving.corr(RQ3_3['Average Clean Sheets/Game']).round(5).astype(str)
```

## Mental Attributes (eg. Agression) vs Cards

```python
RQ3_4 = pd.read_csv(('../../data/analysis/RQ3_4.csv'))
RQ3_4 = RQ3_4 .drop('Unnamed: 0', 1)
RQ3_4 = RQ3_4 .iloc[1:]
```

```python
RQ3_4.head(5)
```

```python
aggression = RQ3_4.groupby(['aggression']).mean()['Average Cards/Game']
phy = RQ3_4.groupby(['phy']).mean()['Average Cards/Game']
strength = RQ3_4.groupby(['strength']).mean()['Average Cards/Game']
sliding_tackle = RQ3_4.groupby(['sliding_tackle']).mean()['Average Cards/Game']
standing_tackle = RQ3_4.groupby(['standing_tackle']).mean()['Average Cards/Game']
marking = RQ3_4.groupby(['marking']).mean()['Average Cards/Game']
```

```python
aggression_corr = aggression.corr(RQ3_4['Average Cards/Game']).round(5).astype(str)
phy_corr = phy.corr(RQ3_4['Average Cards/Game']).round(5).astype(str)
strength_corr = strength.corr(RQ3_4['Average Cards/Game']).round(5).astype(str)
sliding_tackle_corr = sliding_tackle.corr(RQ3_4['Average Cards/Game']).round(5).astype(str)
standing_tackle_corr = standing_tackle.corr(RQ3_4['Average Cards/Game']).round(5).astype(str)
marking_corr = marking.corr(RQ3_4['Average Cards/Game']).round(5).astype(str)
```

```python
axis_size=14
color = ['r']

fig = plt.figure(figsize=(30,15))

gs = gridspec.GridSpec(2,6)
ax0 = plt.subplot(gs[0,0:2],title='\n \n'+ 'Correlation: ' + phy_corr)
ax0.title.set_fontsize(20)
ax0.set_ylabel('Average Cards/Game',size=axis_size)
ax0.set_xlabel('Average Cards/Game',size=axis_size)
phy.plot(ax=ax0,kind='line',color=color)
ax1 = plt.subplot(gs[0,2:4],title='\n \n'+ 'Correlation: ' + aggression_corr)
ax1.title.set_fontsize(20)
ax1.set_ylabel('Average Cards/Game',size=axis_size)
ax1.set_xlabel('Average Cards/Game',size=axis_size)
aggression.plot(ax=ax1,kind='line',color=color)

fig = plt.figure(figsize=(30,15))
gs = gridspec.GridSpec(2,6)
ax0 = plt.subplot(gs[0,0:2],title='\n \n'+ 'Correlation: ' + standing_tackle_corr)
ax0.title.set_fontsize(20)
ax0.set_ylabel('Average Cards/Game',size=axis_size)
ax0.set_xlabel('Average Cards/Game',size=axis_size)

standing_tackle.plot(ax=ax0,kind='line',color=color)
ax1 = plt.subplot(gs[0,2:4],title='\n \n'+ 'Correlation: ' + strength_corr)
ax1.title.set_fontsize(20)
ax1.set_ylabel('Average Cards/Game',size=axis_size)
ax1.set_xlabel('Average Cards/Game',size=axis_size)
strength.plot(ax=ax1,kind='line',color=color)

fig = plt.figure(figsize=(30,15))
gs = gridspec.GridSpec(2,6)
ax0 = plt.subplot(gs[0,0:2],title='\n \n'+ 'Correlation: ' + sliding_tackle_corr)
ax0.title.set_fontsize(20)
ax0.set_ylabel('Average Cards/Game',size=axis_size)
ax0.set_xlabel('Average Cards/Game',size=axis_size)
sliding_tackle.plot(ax=ax0,kind='line',color=color)

ax1 = plt.subplot(gs[0,2:4],title='\n \n'+ 'Correlation: ' + marking_corr)
ax1.title.set_fontsize(20)
ax1.set_ylabel('Average Cards/Game',size=axis_size)
ax1.set_xlabel('Average Cards/Game',size=axis_size)
marking.plot(ax=ax1,kind='line',color=color)
```
