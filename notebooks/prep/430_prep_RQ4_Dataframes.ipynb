# 430_prep_RQ4_Dataframes

```python
import math
import os.path
import numpy as np
import pandas as pd
```

### Loading

```python
DF1 = pd.read_csv("../../data/prep/Joined1617.csv")
DF2 = pd.read_csv("../../data/prep/Joined1516.csv")
DF3 = pd.read_csv("../../data/prep/Joined1415.csv")
DF4 = pd.read_csv("../../data/prep/Joined1314.csv")
DF5 = pd.read_csv("../../data/prep/Joined1213.csv")
```

```python
index = ['Players','club','league','age','nationality','Position','overall','aggression','sliding_tackle','diver_trait','work_rate_def','tackling_speciality','argues_with_officials_trait']
```

```python
s1 = DF1.drop_duplicates(index).set_index(index)[['Apps','Yellow','Red']]
s2 = DF2.drop_duplicates(index).set_index(index)[['Apps','Yellow','Red']]
s3 = DF3.drop_duplicates(index).set_index(index)[['Apps','Yellow','Red']]
s4 = DF4.drop_duplicates(index).set_index(index)[['Apps','Yellow','Red']]
s5 = DF5.drop_duplicates(index).set_index(index)[['Apps','Yellow','Red']]
```

```python
RQ4 = pd.concat([s1,s2,s3,s4,s5], axis=1, keys=('16/17','15/16','14/15','13/14','12/13')).fillna(0).astype(float).reset_index()
```

### Creating New Columns

```python
RQ4
```

```python
RQ4['Total Apps'] = RQ4['16/17']['Apps'] + RQ4['15/16']['Apps'] + RQ4['14/15']['Apps'] + RQ4['13/14']['Apps'] + RQ4['12/13']['Apps']
```

```python
RQ4['Total Yellows'] = RQ4['16/17']['Yellow']+RQ4['15/16']['Yellow']+RQ4['14/15']['Yellow']+RQ4['13/14']['Yellow']+RQ4['12/13']['Yellow']
```

```python
RQ4['Average Yellows/Game'] = RQ4['Total Yellows']/RQ4['Total Apps']
```

```python
RQ4['Total Reds'] = RQ4['16/17']['Red']+RQ4['15/16']['Red']+RQ4['14/15']['Red']+RQ4['13/14']['Red']+RQ4['12/13']['Red']
```

```python
RQ4['Average Reds/Game'] = RQ4['Total Reds']/RQ4['Total Apps']
```

```python
RQ4['Total Cards'] = RQ4['Total Yellows']+RQ4['Total Reds']
```

```python
RQ4['Likelyhood to be Carded'] = (RQ4['Total Reds']+RQ4['Total Yellows'])/RQ4['Total Apps']
```

### Filtering

```python
RQ4 = RQ4[RQ4['Total Apps']>10]
```

```python
 RQ4 = RQ4[RQ4['Total Yellows']>0]
```

### Tidying Up

```python
RQ4.columns = [col[0] for col in RQ4.columns]
```

```python
list(RQ4)
```

```python
RQ4 = RQ4[['Players','club','league','overall','age','Position','aggression','sliding_tackle','diver_trait','work_rate_def','tackling_speciality','argues_with_officials_trait','Total Apps','Total Yellows','Average Yellows/Game','Total Reds','Average Reds/Game','Total Cards','Likelyhood to be Carded']]
```
