```python
import os.path
import numpy as np
import pandas as pd
```

```python
BPL1415 = pd.read_html("https://www.statbunker.com/competitions/FantasyFootballPlayersStats?comp_id=481")[0]
BUN1415 = pd.read_html("https://www.statbunker.com/competitions/FantasyFootballPlayersStats?comp_id=483")[0]
LAL1415 = pd.read_html("https://www.statbunker.com/competitions/FantasyFootballPlayersStats?comp_id=485")[0]
```

```python
SEA1415 = [BPL1415,BUN1415,LAL1415]
SEA1415 = pd.concat(SEA1415)
SEA1415 = SEA1415.reset_index()
```

```python
SEA1415.head()
```

```python
SEA1415.drop(SEA1415.columns[[0,20]], axis=1, inplace=True)
```

```python
SEA1415.head()
```

```python
SEA1415.isnull().sum()
```

```python
DashColumns = ['Goals','A','CS','CS part','Yellow','Red','Sub','CO','Off','Pen SV','Pen M','Goals conceded','Conceded 1+','OG',]
SEA1415[DashColumns] = SEA1415[DashColumns].replace('-' ,'0')
```

```python
SEA1415.info()
```

```python
SEA1415['CS'] = SEA1415['CS'].str.replace('%','').astype(np.int)
SEA1415['CS part'] = SEA1415['CS part'].str.replace('%','').astype(np.int)
SEA1415['Pen SV'] = SEA1415['Pen SV'].str.replace('%','').astype(np.int)
SEA1415['Goals conceded'] = SEA1415['Goals conceded'].str.replace('%','').astype(np.int)
SEA1415['Conceded 1+'] = SEA1415['Conceded 1+'].str.replace('%','').astype(np.int)
```

```python
SEA1415.loc[SEA1415['Players'].str.split().str.len() == 1, 'FirstInitial'] = SEA1415['Players'].astype(str).str[0]
SEA1415.loc[SEA1415['Players'].str.split().str.len() == 1, 'Firstname'] = SEA1415['Players'].str.split().str[0]
SEA1415.loc[SEA1415['Players'].str.split().str.len() == 1, 'Surname'] = SEA1415['Players'].str.split().str[0]

SEA1415.loc[SEA1415['Players'].str.split().str.len() == 2, 'FirstInitial'] = SEA1415['Players'].astype(str).str[0]
SEA1415.loc[SEA1415['Players'].str.split().str.len() == 2, 'Firstname'] = SEA1415['Players'].str.split().str[0]
SEA1415.loc[SEA1415['Players'].str.split().str.len() == 2, 'Surname'] = SEA1415['Players'].str.split().str[-1]

SEA1415.loc[SEA1415['Players'].str.split().str.len() == 3, 'FirstInitial'] = SEA1415['Players'].astype(str).str[0]
SEA1415.loc[SEA1415['Players'].str.split().str.len() == 3, 'Firstname'] = SEA1415['Players'].str.split().str[0]
SEA1415.loc[SEA1415['Players'].str.split().str.len() == 3, 'Surname'] = SEA1415['Players'].str.split().str[-1]

SEA1415.loc[SEA1415['Players'].str.split().str.len() == 4, 'FirstInitial'] = SEA1415['Players'].astype(str).str[0]
SEA1415.loc[SEA1415['Players'].str.split().str.len() == 4, 'Firstname'] = SEA1415['Players'].str.split().str[0]
SEA1415.loc[SEA1415['Players'].str.split().str.len() == 4, 'Surname'] = SEA1415['Players'].str.split().str[-1]
```

```python
SEA1415.set_index(['Surname','FirstInitial'])
```

```python
SEA1415.to_csv('../../data/prep/FANTASY1415.csv')
```
